```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('cleaned_data.csv')
```

```python
df.head()
```

```python
genre_values=df.iloc[:,1:].sum(axis=0)
genre_labels=df.columns[1:]
```

```python
cmap=plt.get_cmap('tab20')
colors=cmap(np.arange(len(genre_values)))
```

```python
fig,ax=plt.subplots(figsize=(10,10),dpi=100)
ax.pie(genre_values,labels=genre_labels,autopct="%1.1f%%",pctdistance=0.8,colors=colors)
plt.show()
```
