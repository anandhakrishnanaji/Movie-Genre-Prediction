```python
import pandas as pd
import numpy as np
import json
```

```python
df=pd.read_csv('../input/movie.metadata.tsv',sep='\t',header=None)
```

df.head()

```python
df_modified=df[[0,2,6,8]]
```

```python
def set_language(x):
    x[6]=list(json.loads(x[6]).values())
    return x
```

```python
df_modified=df_modified.apply(set_language,axis=1)
```

```python
df_modified=df_modified[df_modified.apply(lambda x: bool(x[6]),axis=1)]
```

```python
df_modified.loc[:,6]=df_modified.loc[:,6].apply(lambda x:x[0].split(' ')[0])
```

```python
df_modified
```

```python
df_modified[6].value_counts()
```

```python
df_modified[df_modified[6]=='Malayalam']
```

```python
val=dict(df_modified[6].value_counts())
value_counts_for_charts={'Other':0}
```

```python
for i in val:
    if(val[i]<1000):
        value_counts_for_charts['Other']+=val[i]
    else:
        value_counts_for_charts[i]=val[i]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig=plt.figure(figsize=(17,9))
value_counts_for_charts = dict(sorted(value_counts_for_charts.items(), key=lambda item: item[1],reverse=True))
plt.bar(value_counts_for_charts.keys(),value_counts_for_charts.values())
```

```python
def set_genre(x):
    x[8]=list(json.loads(x[8]).values())
    return x
```

```python
def delete_item(x,string):
    if string in x:
        x.remove(string)
        x.append('Scifi')
#         x.append('Family')
    return x
```
